```python
import numpy as np
from os import listdir
from os.path import isfile, join

mypath = '/content/drive/MyDrive/behavioral_data/RDMsAll/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

struct = {
    'RDM': [],
    'name': [],
    'color': [],
    'sqformRDM': []
        }
RDM = []
name = []
sqfromRDM = []
models = []
for e in onlyfiles: 
    file = open(mypath + e)
    numpy_array = np. loadtxt(file, delimiter=",")
    # print(numpy_array)
    struct['RDM'].append(numpy_array)
    struct['name'].append(e.split(".")[0])
    struct['sqformRDM'].append(numpy_array[np.triu_indices(80, k = 1)])
#     arr = np.array([numpy_array, e.split(".")[0], [], numpy_array[np.triu_indices(3, k = 1)]])
    models.append(struct)
```
